```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['airline'] = le.fit_transform(data['airline'])
data['source_city'] = le.fit_transform(data['source_city'])
data['departure_time'] = le.fit_transform(data['departure_time'])
data['stops'] = le.fit_transform(data['stops'])
data['arrival_time'] = le.fit_transform(data['arrival_time'])
data['destination_city'] = le.fit_transform(data['destination_city'])
data['class'] = le.fit_transform(data['class'])
```

```python
# calculating vif
from statsmodels.stats.outliers_influence import variance_inflation_factor
data1 = data.drop(['price','flight'],axis=1)
vif_data = pd.DataFrame()
vif_data['feature'] = data1.columns
vif_data['vif'] = [variance_inflation_factor(data1.values,i) for i in range(len(data1.columns))]
print(vif_data)
```

```python
# all the vif values are less than 5 hence no need to do anything applying standardization and building linear regression model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = data.drop(['price','flight'],axis=1)
y = data['price']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)

performance_measure(y_test,y_pred)
plotting_graphs(y_test,y_pred)
```

```python
from sklearn.metrics import *

def performance_measure(y_test,y_pred):
  print('MSE : ',mean_squared_error(y_test,y_pred))
  print('RMSE : ', np.sqrt(mean_squared_error(y_test,y_pred)))
  print('mean absolute percentage error MAPE : ',mean_absolute_percentage_error(y_test,y_pred))
  print('mean absolute error MAE : ',mean_absolute_error(y_test,y_pred))
  print('r2 score : ',r2_score(y_test,y_pred))
```

```python
# plotting actual values and predicted values
def plotting_graphs(y_test,y_pred):
  sns.histplot(y_test,label='Actual',kde=True,color='red')
  sns.histplot(y_pred,label='Predicted',kde=True,color='blue')
  plt.legend()
  plt.show()
```

```python
# decision tree regressor
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)
performance_measure(y_test,y_pred)
plotting_graphs(y_test,y_pred)
```

```python
# random forest regressor
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred = rf.predict(x_test)
performance_measure(y_test,y_pred)
plotting_graphs(y_test,y_pred)
```
